```python
import sys
from pathlib import Path

# Try several methods to find the project root so `from src...` works in notebooks
project_root = None

# 1) If running from a real file, use __file__ to locate notebook path (rare in notebooks)
try:
    nb_path = Path(__file__).resolve()
except NameError:
    nb_path = None

if nb_path is not None:
    for p in [nb_path.parent] + list(nb_path.parent.parents):
        if (p / 'src').exists() or (p / 'README.md').exists():
            project_root = p
            break

# 2) Fallback: walk up from the current working directory
if project_root is None:
    cwd = Path.cwd()
    for p in [cwd] + list(cwd.parents):
        if (p / 'src').exists() or (p / 'README.md').exists():
            project_root = p
            break

# 3) Final fallback: explicit known workspace path (adjusted for this environment)
if project_root is None:
    potential = Path('/Users/munawarali/Documents/GitHub/BSK_ODE/Quant-Finance-Summer-2026')
    if potential.exists():
        project_root = potential

# 4) Last-resort: if notebook is inside `experiments`, use its parent; otherwise use cwd
if project_root is None:
    project_root = cwd.parent if 'cwd' in locals() and cwd.name == 'experiments' else (cwd if 'cwd' in locals() else Path('.').resolve())

# Insert at front so local `src` takes precedence
if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))

print("Added to sys.path:", project_root)
```
